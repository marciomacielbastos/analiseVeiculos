### Como não consegui extrair o href referente às URLs a serem visitadas diretamente, criei essa gambiarra usando o BeautifulSoup

```python
def takeNext(d):#Gambiarra
    sopa=soup(d.page_source.encode('utf-8'))
    k=[]
    for a in sopa.findAll('a', href=True):
        k.append(a['href'])
    return k
```

### Método para extrair os dados de interesse da HTML minerada

```python
def mineraCarro(tabela,S,j,i):
    #A posicao "j" do array foi passada separadamente para possível correção de erros provocados pela página visitada
    if j>=40:
        return tabela,j,0
    driver.get(S[j])
    try:
        tabela['Marca'][i] = driver.find_element_by_xpath('//*[@id="description-text"]/div/dl[1]/dd[1]').text
        tabela['Modelo'][i] = driver.find_element_by_xpath('//*[@id="description-text"]/div/dl[1]/dd[2]').text
        tabela['Valor'][i] = int(float(driver.find_element_by_xpath('//*[@id="sticky-container"]/div/div/span').text[3:].replace('.','')))
        tabela['Ano'][i] = int(driver.find_element_by_xpath('//*[@id="description-text"]/div/dl[2]/dd[1]').text)
    except: 
        j=j+1
        mineraCarro(tabela,S,j,i)
    time.sleep(5)
    return tabela,j,1
```

### Método para navegação e criação de array de URLs de interesse

```python
def mineraPagina(S,tabela,posicao):
    S1=S
    j=0
    flag =1
    for i in range(30):
        j=0
        driver.get(S)
        address=takeNext(driver)
        #Extração de somente 40 URLs por página de exibição
        address=address[63:104]
        while j<40:
            tabela,j,flag=mineraCarro(tabela,address,j,posicao)
            posicao+=1
            j+=1
            if flag==0:
                break
        
        num=i+2
        S=S1+'-p-'+str(num)
        if posicao==len(tabela):
            break
    return tabela
```
